```python
df['bathrooms_numeric'] = df['bathrooms_text'].str.extract(r'(\d+(\.\d+)?)')[0]
df['bathrooms_numeric'] = pd.to_numeric(df['bathrooms_numeric'], errors='coerce')
df['shared_bathroom'] = df['bathrooms_text'].apply(lambda x: 'Yes' if isinstance(x, str) and 'shared' in x.lower() else 'No')
print(df[['bathrooms_text', 'bathrooms_numeric', 'shared_bathroom']].head(10))

```

Listing / Հյուրընկալվող բնակարանները

id → Յուրաքանչյուր բնակարանի

name → Անունը

neighbourhood_cleansed → Թարմացված թաղամասի անունը (օրինակ, Shibuya, Shinjuku)

price → Ցուցակի գինը Յենով (¥)

latitude → Շրջանի աշխարհագրական լայնություն

longitude → Շրջանի աշխարհագրական երկայնություն

property_type → Հարկի տիպը ( բնակարան, սենյակ և այլն)

room_type → Սենյակի տիպը (entire home/apartment, private room, shared room)

accommodates → Քանի մարդ կարող է տեղավորվել

bedrooms → Սենյակների քանակը

beds → Մահճակալների քանակը

bathrooms_text → Կողմնակի բաղնիքի նկարագրություն (private/shared, 1, 1.5, 2)

amenities → Հարմարություններ ցուցակի մեջ (WiFi, TV, Air conditioning և այլն)

minimum_nights → Նվազագույն մնալու գիշերների քանակը

maximum_nights → Վերցնելու առավելագույն գիշերների քանակը

Availability / Հասանելիություն

availability_30 → Հասանելի օրերի քանակը հաջորդ 30 օրվա ընթացքում

availability_365 → Հասանելի օրերի քանակը հաջորդ 365 օրվա ընթացքում

has_availability → Ցուցակը արդյոք հասանելի է (Yes/No)

Reviews ֊ Քննադատություններ/կարծիքներ

number_of_reviews → Ընդհանուր ստացված կարծիքների քանակը

number_of_reviews_l30d → Վերջին 30 օրվա կարծիքների քանակը

review_scores_rating → Ընդհանուր գնահատական (1–10)

review_scores_cleanliness → Մաքրության գնահատական

review_scores_location → Գտնվելու վայրի գնահատական

reviews_per_month → Ամսական գնահատականների միջին քանակը

Host / Հյուրընկալող

host_id → Յուրաքանչյուր հյուրընկալողի

host―name → Հյուրընկալողի անունը

host_is_superhost → Հյուրընկալողը Superhost է (Yes/No)

host_listings_count → Հյուրընկալողի ցուցակների քանակը տվյալ գծում (սովորաբար մեկի մասին)

host_total_listings_count → Հյուրընկալողի բոլոր ցուցակների քանակը

host_identity_verified → Հյուրընկալողի նույնականացումը հաստատված է (Yes/No)

Այս հետազոտական տվյալների վերլուծության ընթացքում ուսումնասիրվեց Tokyo Airbnb 2023 տվյալների հավաքածուն՝ նպատակ ունենալով բացահայտել գների ձևավորման օրինաչափությունները, բնակարանների բնութագրերը և թաղամասերի միջև տարբերությունները։ Վերլուծությունը ցույց տվեց, որ Airbnb բնակարանների գները Տոկիոյում էապես տարբերվում են՝ կախված տեղակայությունից, բնակարանի տեսակից և տեղավորման կարողությունից։

Վերլուծության արդյունքում պարզ դարձավ, որ քաղաքի կենտրոնին և զբոսաշրջային տարածքներին մոտ գտնվող բնակարաններն ունեն ավելի բարձր գներ, մինչդեռ մասնավոր և համատեղ սենյակները սովորաբար ավելի մատչելի են ամբողջական բնակարանների համեմատ։ Բացի այդ, գնահատականների (reviews) քանակը և տարվա ընթացքում հասանելիությունը կարևոր պատկերացում տվեցին պահանջարկի և ակտիվության վերաբերյալ։

Տվյալների հավաքածուում առկա էին բացակայող արժեքներ և արտաբերվող արժեքներ (outliers), հատկապես գների հետ կապված հատկանիշներում, որոնք կարող են ազդել հետագա վերլուծությունների արդյունքների վրա և պահանջում են նախնական մշակում։ Չնայած այս սահմանափակումներին՝ տվյալները բացահայտեցին հստակ միտումներ և նշանակալի կապեր հիմնական հատկանիշների միջև։
